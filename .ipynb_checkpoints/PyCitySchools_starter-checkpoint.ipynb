```python
import pandas as pd

school_data_to_load = "C:\\Users\\me\\Desktop\\BBC\\COPY_but-ind-data-pt-06-2020-u-c\\Unit_04_Pandas\\Homework\\PyCitySchools\\Resources\\schools_complete.csv"
student_data_to_load = "C:\\Users\\me\\Desktop\\BBC\\COPY_but-ind-data-pt-06-2020-u-c\\Unit_04_Pandas\\Homework\\PyCitySchools\\Resources\\students_complete.csv"

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

* Calculate the total number of students

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

```python
school_data_complete["school_name"].unique()
```

```python
totalSchools = school_data_complete["school_name"].unique()
total_school_count = len(totalSchools)
```

```python
totalKids = school_data_complete["student_name"].count()
```

```python
totalBudget = school_data["budget"].sum()
```

```python
averageMath = school_data_complete["math_score"].mean() 
```

```python
averageReading = school_data_complete["reading_score"].mean() 
```

```python
mathBins = 0, 69.9, 100
math_names = ["Failing", "Passing"]
Math_df = pd.cut(school_data_complete["math_score"], mathBins, labels=math_names, include_lowest=True)
```

```python
total_mpass_scores = Math_df.value_counts("Passing")
```

```python
readingBins = 0, 69.9, 100
reading_names = ["Failing", "Passing"]
Reading_df = pd.cut(school_data_complete["reading_score"], readingBins, labels=reading_names, include_lowest=True)
```

```python
total_rpass_scores = Reading_df.value_counts("Passing")
```

```python
totalMpass = total_mpass_scores["Passing"]
totalRpass = total_rpass_scores["Passing"]
```

```python
overall = school_data_complete[(school_data_complete["math_score"] >= 70)
                               & (school_data_complete["reading_score"] >= 70)].count()["student_name"]
overallPercent = overall / totalKids 
```

```python
districtSummary = pd.DataFrame({"Total Schools": [total_school_count],
                              "Total Students": [totalKids],
                              "Total Budget": [totalBudget],
                              "Average Math Score": [averageMath],
                               "Average Reading Score": [averageReading],
                               "Math Percentage": [totalMpass],
                                "Reading Percentage": [totalRpass],
                                "Overall Passing Percentage": [overallPercent]
                                  })

format_dict = {'Total Budget': '${0:,.2f}', 'Math Percentage': '{:.2%}', 'Reading Percentage': '{:.2%}', 'Overall Passing Percentage': '{:.2%}'}
districtSummary.style.format(format_dict).hide_index()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
totalStudents = school_data_complete_df.groupby(["school_name"])["student_name"].count()
schoolBudget = school_data_complete_df.groupby(["school_name"])["budget"].mean()
perSchoolBudget = schoolBudget / totalStudents
```

```python
avMath = school_data_complete_df.groupby(["school_name"])["math_score"].mean()
avRead = school_data_complete_df.groupby(["school_name"])["reading_score"].mean()
```

```python
mathScore = school_data_complete[(school_data_complete["math_score"] >= 70)]
mathPercent = mathScore.groupby(["school_name"])["math_score"].count() / totalStudents
```

```python
readScore = school_data_complete[(school_data_complete["reading_score"] >= 70)]
readPercent = readScore.groupby(["school_name"])["reading_score"].count() / totalStudents
```

```python
over_all = school_data_complete[(school_data_complete["math_score"] >= 70)
                               & (school_data_complete["reading_score"] >= 70)] #isolated rows with both conditions 
over_allPercent = over_all.groupby(["school_name"])["student_name"].count() / totalStudents
```

```python
s_types = school_data.set_index(["school_name"])["type"]
```

```python
schoolSummary_df = pd.DataFrame({"School Type": s_types,
                                 "Total Students": totalStudents, 
                             "Total School Budget": schoolBudget,
                             "Per Student Budget": perSchoolBudget,
                             "Average Math Score": avMath,
                             "Average Reading Score": avRead,
                             "% Passing Math": mathPercent,
                             "% Passing Reading": readPercent,
                             "% Overall Passing": over_allPercent})                       
schoolSummary_df
```
