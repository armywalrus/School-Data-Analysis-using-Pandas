* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

```python
school_data_complete["School Name"].unique()
```

```python
total_schools = school_data_complete["School Name"].unique()
total_school_count = len(total_schools)
```

```python
total_kids = school_data_complete["student_name"].count()
```

```python
total_budget = school_data["budget"].sum()
```

```python
average_math = school_data_complete["math_score"].mean()
average_reading = school_data_complete["reading_score"].mean()
```

```python
math_bins = 0, 69.9, 100
math_names = ["Failing", "Passing"]
Math_df = pd.cut(school_data_complete["math_score"], math_bins, labels=math_names, include_lowest=True)
total_mpass_scores = Math_df.value_counts("Passing")
```

```python
reading_bins = 0, 69.9, 100
reading_names = ["Failing", "Passing"]
Reading_df = pd.cut(school_data_complete["reading_score"], reading_bins, labels=reading_names, include_lowest=True)
total_rpass_scores = Reading_df.value_counts("Passing")
```

```python
total_math_pass = total_mpass_scores["Passing"]
total_reading_pass = total_rpass_scores["Passing"]
```

```python
overall = school_data_complete[(school_data_complete["math_score"] >= 70)
                               & (school_data_complete["reading_score"] >= 70)].count()["student_name"]
overall_percent = overall / total_kids
```

```python
district_summary = pd.DataFrame({"Total Schools": [total_school_count],
                              "Total Students": [total_kids],
                              "Total Budget": [total_budget],
                              "Average Math Score": [average_math],
                               "Average Reading Score": [average_reading],
                               "% Passing Math": [total_math_pass],
                                "% Passing Reading": [total_reading_pass],
                                "% Overall Passing": [overall_percent]})       
format_dict = {'Total Budget': '${0:,.2f}', 'Total Students': '{:,.0f}', 'Average Math Score': '{:.2f}', 'Average Reading Score': '{:.2f}', 
                 '% Passing Math': '{:.2%}', '% Passing Reading': '{:.2%}', '% Overall Passing': '{:.2%}'}

district_summary.style.format(format_dict).hide_index()
```

```python
overall_score = school_data_complete[(school_data_complete["math_score"] >= 70)
                               & (school_data_complete["reading_score"] >= 70)] 
overall_percent = overall_score.groupby(["School Name"])["student_name"].count() / total_students
```
